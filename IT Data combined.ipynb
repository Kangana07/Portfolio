```python
import pandas as pd
```

```python
data2018=pd.read_csv(r"C:\Users\BAPS\Downloads\2018_IT_cleaned.csv")
```

```python
data2019=pd.read_csv(r"C:\Users\BAPS\Downloads\2019_IT_cleaned.csv")
```

```python
data2020=pd.read_csv(r"C:\Users\BAPS\Downloads\2020_IT_cleaned.csv")
```

```python
print(data2018.columns)
```

```python
print(data2019.columns)
```

```python
print(data2020.columns)
```

```python
# Align columns first by creating empty placeholders for missing columns
all_columns = list(set(data2018.columns).union(set(data2019.columns), set(data2020.columns)))

data2018 = data2018.reindex(columns=all_columns, fill_value=None)
data2019 = data2019.reindex(columns=all_columns, fill_value=None)
data2020 = data2020.reindex(columns=all_columns, fill_value=None)

# Concatenate after alignment
combined_data = pd.concat([data2018, data2019, data2020], ignore_index=True)
```
