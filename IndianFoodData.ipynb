```python
data['region'] = data['region'].fillna('unknown')
```

```python
# Convert the specified columns to string to ensure '-1' is matched
columns_to_check = ['flavor_profile', 'state', 'region']
data[columns_to_check] = data[columns_to_check].astype(str)

# Drop rows where '-1' exists in the specified columns
data = data[~data[columns_to_check].isin(['-1']).any(axis=1)]

# Verify if '-1' is removed
print(data[columns_to_check].isin(['-1']).any())


```
